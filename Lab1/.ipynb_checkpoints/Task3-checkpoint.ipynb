## Постановка задачи

Рассматривается **автономный беспроводной сенсор** с солнечной батареей, предназначенный для проведения исследований. Необходимо определить оптимальную стратегию передачи данных, максимизирующую ожидаемое количество доставленных пакетов данных.

### Описание модели

- **Дискретное время**: процесс разбит на $T$ временных слотов.
- **Управление**: в каждый слот сенсор может **передавать данные** ($u=1$) или **ожидать** ($u=0$).
- **Передача данных**: требует 1 единицу энергии.
  - В **благоприятном** состоянии канала: сообщение доставляется с вероятностью $p_{\text{good}} = 0.99$.
  - В **неблагоприятном** состоянии канала: сообщение доставляется с вероятностью $p_{\text{bad}} = 0.45$.
- **Энергия**: каждый слот — вероятность получения 1 единицы от солнечной батареи $p_{\text{solar}} = 0.2$. Ёмкость батареи: $E_{\max} = 20$.
- **Динамика канала** (марковская цепь с двумя состояниями):
  - $P(0 \to 1) = 0.2$, $P(0 \to 0) = 0.8$ (неблагоприятный → благоприятный / неблагоприятный)
  - $P(1 \to 1) = 0.7$, $P(1 \to 0) = 0.3$ (благоприятный → благоприятный / неблагоприятный)

### Задания

1. Найти ожидаемое количество доставленных пакетов при оптимальной стратегии за $T = 5000$ слотов.
2. Сравнить оптимальные стратегии для $T = 1000$ и $T = 2000$: есть ли различия и в чём они состоят.
3. Промоделировать реализацию случайного процесса, порождённого оптимальной стратегией.
4. Исследовать зависимость ожидаемого числа доставленных пакетов от вероятности подзарядки $p_{\text{solar}}$.

## Почему это задача стохастического динамического программирования

Данная задача является **марковским процессом принятия решений** (MDP) с конечным горизонтом, что представляет собой классическую модель **стохастического динамического программирования**.

В отличие от детерминированных задач ДП (таких как задача о рюкзаке или о распределении ресурсов), в нашей задаче присутствуют **два источника случайности**:

1. **Случайные переходы канала связи** — канал переключается между благоприятным и неблагоприятным состояниями по марковской цепи, и это происходит *независимо* от действий сенсора.
2. **Случайное поступление энергии** — солнечная батарея с вероятностью $p_{\text{solar}} = 0.2$ пополняет запас на 1 единицу.

Таким образом, при выборе управления $u_i$ в состоянии $(e_i, c_i)$ мы **не можем точно предсказать** следующее состояние $(e_{i+1}, c_{i+1})$. Можно лишь определить **распределение вероятностей** перехода.

Тем не менее, задача обладает свойствами, необходимыми для применения принципа ДП:

- **Оптимальная структура подзадач**: оптимальное управление на оставшихся шагах не зависит от того, как именно мы попали в текущее состояние (марковость).
- **Пересекающиеся подзадачи**: для разных траекторий может понадобиться одно и то же значение $W_t(e, c)$.

Аналогия с классической **задачей о садовнике** из курса: там садовник выбирал, вносить удобрения или нет, наблюдая стохастическое состояние почвы. Здесь сенсор выбирает, передавать данные или нет, наблюдая случайное состояние канала и имеющуюся энергию. Состояние включает как **управляемую** компоненту (запас энергии, меняющийся в зависимости от решения передавать) так и **неуправляемую** (состояние канала, определяемое внешними факторами).

## Построение модели

Следуя стандартному алгоритму моделирования задач ДП (аналогично описанному в примерах курса), выделим основные компоненты модели.

### 1. Пространство состояний

Состояние системы на шаге $t$ описывается парой:

$$
s_t = (e_t, c_t),
$$

где:
- $e_t \in \{0, 1, \ldots, 20\}$ — текущий уровень заряда батареи (21 значение),
- $c_t \in \{0, 1\}$ — состояние канала связи: $0$ — неблагоприятное, $1$ — благоприятное (2 значения).

Итого: $|S| = 21 \times 2 = 42$ состояния.

### 2. Пространство управлений

$$
u_t \in U(e_t) = \begin{cases} \{0\}, & \text{если } e_t = 0, \\ \{0, 1\}, & \text{если } e_t \geq 1. \end{cases}
$$

- $u_t = 0$: ожидать (не передавать данные),
- $u_t = 1$: передавать данные (расходуется 1 единица энергии).

### 3. Немедленный выигрыш (награда)

Выигрыш определяется как **ожидаемое количество доставленных пакетов** в данном слоте:

$$
r(u, c) = \begin{cases} p_{\text{good}} = 0.99, & \text{если } u = 1 \text{ и } c = 1, \\ p_{\text{bad}} = 0.45, & \text{если } u = 1 \text{ и } c = 0, \\ 0, & \text{если } u = 0. \end{cases}
$$

### 4. Функция перехода

В стохастической задаче переход задаётся не функцией, а **распределением вероятностей**.

**Энергия:**
- Базовый уровень: $e_{\text{base}} = e_t - u_t$ (расход энергии при передаче).
- Затем поступление от солнечной батареи: $\xi_t \sim \text{Bernoulli}(p_{\text{solar}})$.
- Новый уровень: $e_{t+1} = \min(e_{\text{base}} + \xi_t, E_{\max})$.

**Канал (марковская цепь, независимо от управления):**

$$
P_{\text{channel}} = \begin{pmatrix} 0.8 & 0.2 \\ 0.3 & 0.7 \end{pmatrix}, \quad
P_{\text{channel}}[c, c'] = P(c_{t+1} = c' \mid c_t = c).
$$

## Уравнение Беллмана

Центральный элемент решения — **уравнение Беллмана** (основное функциональное уравнение стохастического ДП). В отличие от детерминированного случая, здесь под знаком максимума стоит **математическое ожидание** по случайным переходам:

$$
\boxed{
W_t(e, c) = \max_{u \in U(e)} \left\{ \sum_{c' \in \{0,1\}} P_{\text{channel}}(c'|c) \sum_{\xi \in \{0,1\}} P(\xi) \left[ r(u, c) + W_{t+1}\left(\min(e - u + \xi, \, E_{\max}), \, c'\right) \right] \right\}
}
$$

где:
- $W_t(e, c)$ — **условный оптимальный ожидаемый выигрыш** (максимальное ожидаемое число доставленных пакетов с шага $t$ до конца горизонта), если к шагу $t$ система находится в состоянии $(e, c)$;
- $U(e) = \{0\}$ при $e = 0$, иначе $\{0, 1\}$;
- $r(u, c)$ — немедленный выигрыш (вероятность успешной передачи);
- $P_{\text{channel}}(c'|c)$ — вероятность перехода канала из состояния $c$ в $c'$;
- $P(\xi = 1) = p_{\text{solar}} = 0.2$, $P(\xi = 0) = 0.8$;
- $E_{\max} = 20$;
- **Граничное условие**: $W_T(e, c) = 0$ для всех $(e, c)$ (после горизонта выигрыша нет).

Решение производится методом **обратной прогонки** (backward induction): начиная с последнего шага $T-1$ и двигаясь к шагу $0$, последовательно вычисляем $W_t(e, c)$ и запоминаем оптимальное управление $u^*_t(e, c)$.

## Реализация

### Параметры задачи

Определим все числовые параметры модели.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Настройка графиков
plt.rcParams.update({
    'figure.figsize': (12, 5),
    'figure.dpi': 120,
    'font.size': 11,
    'axes.titlesize': 13,
    'axes.titleweight': 'bold',
    'axes.labelsize': 11,
})

# --- Параметры задачи ---

# Вероятности доставки пакета
p_good = 0.99   # в благоприятном канале
p_bad = 0.45    # в неблагоприятном канале

# Энергия
p_solar = 0.2   # вероятность получения 1 единицы энергии
E_max = 20      # ёмкость батареи

# Матрица переходов канала: P_ch[c, c']
# c=0: неблагоприятный, c=1: благоприятный
P_ch = np.array([
    [0.8, 0.2],   # из неблагоприятного: остаться 0.8, перейти в благ. 0.2
    [0.3, 0.7]    # из благоприятного: перейти в неблаг. 0.3, остаться 0.7
])

# Проверка корректности матрицы переходов
assert np.allclose(P_ch.sum(axis=1), 1.0), "Строки матрицы переходов должны суммироваться в 1"

# Функция немедленного выигрыша
def reward(u, c):
    """Ожидаемое число доставленных пакетов при управлении u и состоянии канала c."""
    if u == 0:
        return 0.0
    else:  # u == 1
        return p_good if c == 1 else p_bad

print("Параметры модели:")
print(f"  p_good = {p_good}, p_bad = {p_bad}")
print(f"  p_solar = {p_solar}, E_max = {E_max}")
print(f"  Матрица переходов канала:\n{P_ch}")
print(f"  Число состояний: {(E_max + 1) * 2} = {E_max + 1} (энергия) × 2 (канал)")
```

### Стационарное распределение канала

Для усреднения начальных условий полезно знать стационарное распределение марковской цепи канала. Решаем $\pi P_{\text{channel}} = \pi$, $\sum \pi_i = 1$:

```python
# Стационарное распределение канала
# π P = π, сумма π = 1
# P(0→1) = 0.2, P(1→0) = 0.3
# π_0 * 0.2 = π_1 * 0.3  =>  π_0 / π_1 = 3/2  =>  π_0 = 3/5, π_1 = 2/5
pi_0 = 0.3 / (0.2 + 0.3)  # = 0.6
pi_1 = 0.2 / (0.2 + 0.3)  # = 0.4
pi_channel = np.array([pi_0, pi_1])

# Проверка
assert np.allclose(pi_channel @ P_ch, pi_channel), "Ошибка: не стационарное распределение"
print(f"Стационарное распределение канала: π₀={pi_0:.2f} (неблаг.), π₁={pi_1:.2f} (благ.)")
print(f"Канал находится в благоприятном состоянии примерно {pi_1*100:.0f}% времени.")
```

### Обратная прогонка (backward induction)

Ключевой шаг решения — вычисление уравнения Беллмана методом обратной прогонки. Начиная с граничного условия $W_T(e, c) = 0$, последовательно двигаемся от $t = T-1$ к $t = 0$.

Для эффективности используем массивы numpy размера $(E_{\max}+1, 2)$ для текущего и следующего шага. Полную стратегию храним в массиве `policy[t, e, c]`.

**Важное наблюдение:** при большом $T$ оптимальная стратегия быстро стабилизируется (стационарная стратегия). Различия наблюдаются лишь вблизи конца горизонта (последние ~30–40 шагов).

```python
def solve_bellman(T, p_solar_val=0.2, return_policy=False, return_all_W=False):
    """
    Решение уравнения Беллмана методом обратной прогонки.
    
    Параметры:
        T             - горизонт планирования (число временных слотов)
        p_solar_val   - вероятность подзарядки
        return_policy - если True, возвращает полную стратегию policy[t, e, c]
        return_all_W  - если True, возвращает W для всех t
    
    Возвращает:
        W0            - значения функции Беллмана на шаге 0: W0[e, c]
        policy        - (опционально) оптимальная стратегия: policy[t, e, c]
        W_all         - (опционально) все значения: W_all[t, e, c]
    """
    n_energy = E_max + 1  # 0..20
    n_channel = 2         # 0, 1
    
    # Граничное условие: W_T = 0
    W_next = np.zeros((n_energy, n_channel))
    
    # Массив стратегии (если запрошен)
    if return_policy:
        policy = np.zeros((T, n_energy, n_channel), dtype=np.int8)
    if return_all_W:
        W_all = np.zeros((T + 1, n_energy, n_channel))
        W_all[T] = 0.0
    
    # Вероятности поступления энергии
    p_no_solar = 1.0 - p_solar_val
    
    # Обратная прогонка: от t = T-1 до t = 0
    for t in range(T - 1, -1, -1):
        W_curr = np.zeros((n_energy, n_channel))
        
        for e in range(n_energy):
            for c in range(n_channel):
                best_val = -np.inf
                best_u = 0
                
                # Перебор допустимых управлений
                controls = [0, 1] if e >= 1 else [0]
                
                for u in controls:
                    # Немедленный выигрыш
                    r = reward(u, c)
                    
                    # Базовый уровень энергии после действия
                    e_base = e - u
                    
                    # Математическое ожидание по случайным переходам
                    expected_future = 0.0
                    for c_new in range(2):
                        p_c = P_ch[c, c_new]
                        for xi in range(2):  # xi=0 или xi=1
                            p_xi = p_solar_val if xi == 1 else p_no_solar
                            e_new = min(e_base + xi, E_max)
                            expected_future += p_c * p_xi * W_next[e_new, c_new]
                    
                    total = r + expected_future
                    
                    if total > best_val:
                        best_val = total
                        best_u = u
                
                W_curr[e, c] = best_val
                if return_policy:
                    policy[t, e, c] = best_u
        
        if return_all_W:
            W_all[t] = W_curr
        W_next = W_curr
    
    result = [W_curr]
    if return_policy:
        result.append(policy)
    if return_all_W:
        result.append(W_all)
    return result[0] if len(result) == 1 else tuple(result)
```

## Задание 1: Ожидаемое число доставленных пакетов при $T = 5000$

Запустим обратную прогонку для $T = 5000$ и определим оптимальное ожидаемое число доставленных пакетов для различных начальных условий.

```python
%%time
# Решение для T = 5000
T_main = 5000
W0_5000, policy_5000 = solve_bellman(T_main, return_policy=True)

print("=" * 60)
print(f"Результаты для T = {T_main}")
print("=" * 60)

# Разные начальные условия
for e_init in [0, 5, 10, 15, 20]:
    for c_init, c_name in [(0, 'неблаг.'), (1, 'благ.')]:
        val = W0_5000[e_init, c_init]
        print(f"  W₀(e={e_init:2d}, c={c_name:8s}) = {val:.2f} пакетов")

print()
# Ожидаемое значение при начальной энергии e и стационарном распределении канала
for e_init in [0, 10, 20]:
    avg = pi_channel[0] * W0_5000[e_init, 0] + pi_channel[1] * W0_5000[e_init, 1]
    print(f"  E[W₀(e={e_init:2d})] (по стац. распр. канала) = {avg:.2f} пакетов")
```

### Интерпретация результатов

Обратим внимание, что:
- Значение $W_0$ слабо зависит от начального уровня энергии (при большом $T$ влияние начальных условий нивелируется за счёт регулярной подзарядки).
- Различие между полной и пустой батареей — порядка нескольких пакетов из тысяч.
- Основной вклад определяется **стационарным режимом** работы стратегии.

### Проверка здравого смысла (sanity check)

Оценим разумность результата. В стационарном режиме:
- Поступает в среднем $0.2$ единиц энергии в слот.
- Частота передачи ограничена скоростью зарядки: не более $\approx 0.2$ передач в слот.
- Средняя вероятность доставки зависит от того, в каком канале передаём.
- Верхний предел: если всегда передавать только в благоприятном канале — $T \cdot p_{\text{solar}} \cdot p_{\text{good}}$.

```python
# Sanity check
avg_delivery_any_channel = pi_0 * p_bad + pi_1 * p_good
print(f"Средняя вероятность доставки (без выбора канала): {avg_delivery_any_channel:.4f}")

# Грубая верхняя оценка: передаём каждый слот (0.2 энергии/слот в среднем)
upper_naive = T_main * p_solar * avg_delivery_any_channel
print(f"Грубая оценка (передача без выбора канала): {upper_naive:.1f}")

# Оптимальная стратегия может быть лучше за счёт выбора канала
upper_good_only = T_main * p_solar * p_good  
print(f"Верхняя оценка (только благ. канал, неогр. энергия): {upper_good_only:.1f}")

# Фактическое значение
avg_actual = pi_channel @ W0_5000[10, :]
print(f"\nОптимальное W₀(e=10, стац. распр.) = {avg_actual:.2f}")
print(f"Это примерно {avg_actual / T_main:.4f} пакетов в слот")
print(f"\nОптимальная стратегия лучше 'наивной' на ~{(avg_actual - upper_naive)/upper_naive*100:.0f}%")
```

### Стационарная оптимальная стратегия

Покажем оптимальную стратегию в стационарном режиме (далеко от конца горизонта). Для каждого уровня энергии и состояния канала указываем оптимальное действие.

```python
# Стационарная стратегия (шаг 0, далеко от конца горизонта при T=5000)
stationary_policy = policy_5000[0, :, :]

print("Стационарная оптимальная стратегия:")
print(f"{'Энергия':>8s}  {'Неблаг.(c=0)':>14s}  {'Благ.(c=1)':>14s}")
print("-" * 40)
for e in range(E_max + 1):
    u0 = 'Передавать' if stationary_policy[e, 0] == 1 else 'Ждать'
    u1 = 'Передавать' if stationary_policy[e, 1] == 1 else 'Ждать'
    print(f"  e={e:2d}      {u0:>14s}  {u1:>14s}")

# Определим пороговые значения
for c, c_name in [(0, 'неблагоприятном'), (1, 'благоприятном')]:
    transmit_e = np.where(stationary_policy[:, c] == 1)[0]
    if len(transmit_e) > 0:
        print(f"\nПорог для {c_name} канала: e* = {transmit_e[0]} "
              f"(передаём при e ≥ {transmit_e[0]})")
    else:
        print(f"\nВ {c_name} канале никогда не передаём")
```

**Ключевое наблюдение**: оптимальная стратегия в стационарном режиме практически никогда не передаёт в **неблагоприятном** канале (кроме случая полной батареи $e = E_{\max} = 20$, где входящая энергия будет потеряна). В **благоприятном** канале сенсор передаёт при любом уровне энергии $e \geq 1$.

Это логично: вероятность доставки в благоприятном канале ($0.99$) более чем вдвое выше, чем в неблагоприятном ($0.45$). При ограниченной энергии ($p_{\text{solar}} = 0.2$) выгоднее «копить» энергию и расходовать её в благоприятном канале.

## Задание 2: Сравнение стратегий для $T = 1000$ и $T = 2000$

Вычислим оптимальные стратегии для обоих горизонтов и сравним их. Ожидается, что стратегии будут совпадать в **стационарной части** (далеко от конца горизонта) и различаться лишь вблизи терминального времени.

```python
%%time
# Решение для T = 1000 и T = 2000
T1, T2 = 1000, 2000

W0_1000, policy_1000 = solve_bellman(T1, return_policy=True)
W0_2000, policy_2000 = solve_bellman(T2, return_policy=True)

print(f"Ожидаемое число пакетов (e=10, стац. распр. канала):")
for T_val, W0 in [(T1, W0_1000), (T2, W0_2000), (T_main, W0_5000)]:
    avg = pi_channel @ W0[10, :]
    print(f"  T={T_val:5d}: {avg:.2f} пакетов ({avg/T_val:.4f} пакетов/слот)")
```

### Анализ сходимости стратегии к стационарной

Определим, на каком расстоянии от конца горизонта стратегия стабилизируется (перестаёт изменяться). Для этого сравним стратегию на каждом шаге с предыдущей, двигаясь от конца к началу.

```python
def find_convergence_step(policy, T_val):
    """Определяет, за сколько шагов до конца стратегия отличается от стационарной."""
    # Стационарная стратегия — в начале горизонта (t=0)
    stationary = policy[0]
    # Ищем от конца: первый шаг, совпадающий со стационарной
    for t in range(T_val - 1, -1, -1):
        if np.array_equal(policy[t], stationary):
            return T_val - t - 1  # последние N шагов отличаются
    return T_val

conv_1000 = find_convergence_step(policy_1000, T1)
conv_2000 = find_convergence_step(policy_2000, T2)
conv_5000 = find_convergence_step(policy_5000, T_main)

print(f"Стратегия отличается от стационарной в последних N шагах горизонта:")
print(f"  T={T1:5d}: нестационарная зона = {conv_1000} шагов")
print(f"  T={T2:5d}: нестационарная зона = {conv_2000} шагов")
print(f"  T={T_main:5d}: нестационарная зона = {conv_5000} шагов")

# Совпадают ли стационарные части?
print(f"\nСтационарные стратегии T={T1} и T={T2} совпадают: "
      f"{np.array_equal(policy_1000[0], policy_2000[0])}")
print(f"Стационарные стратегии T={T2} и T={T_main} совпадают: "
      f"{np.array_equal(policy_2000[0], policy_5000[0])}")
```

### Визуализация: пороговая энергия vs. оставшееся время

Определим **пороговую энергию** $e^*(\tau, c)$ — минимальный уровень энергии, при котором оптимально передавать, как функцию оставшегося времени $\tau = T - t$ и состояния канала $c$. Это наглядно покажет, как стратегия меняется вблизи конца горизонта.

```python
def get_threshold_energy(policy, T_val):
    """Определяет пороговую энергию как функцию оставшегося времени."""
    threshold = np.zeros((T_val, 2))  # [t, c]
    
    for t in range(T_val):
        remaining = T_val - t
        for c in range(2):
            transmit_energies = np.where(policy[t, :, c] == 1)[0]
            if len(transmit_energies) > 0:
                threshold[t, c] = transmit_energies[0]
            else:
                threshold[t, c] = E_max + 1  # никогда не передаём
    
    return threshold

fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

for T_val, policy, color, label in [
    (T1, policy_1000, '#20808D', f'T={T1}'),
    (T2, policy_2000, '#A84B2F', f'T={T2}'),
    (T_main, policy_5000, '#1B474D', f'T={T_main}')
]:
    threshold = get_threshold_energy(policy, T_val)
    remaining = np.arange(T_val, 0, -1)  # от T до 1
    
    for c, c_name in [(0, 'Неблагоприятный канал (c=0)'), (1, 'Благоприятный канал (c=1)')]:
        ax = axes[c]
        # Показываем первые 100 шагов от конца (оставшееся время 1..100)
        n_show = min(100, T_val)
        idx_start = T_val - n_show  # шаги от конца
        ax.plot(remaining[idx_start:], threshold[idx_start:, c],
                label=label, linewidth=1.5, color=color)
        ax.set_title(c_name)
        ax.set_xlabel('Оставшееся время τ = T − t')
        ax.set_ylabel('Пороговая энергия e*')
        ax.legend()
        ax.set_ylim(0, E_max + 2)
        ax.grid(True, alpha=0.3)

plt.suptitle('Пороговая энергия для передачи: зависимость от оставшегося времени',
             fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('threshold_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Задание 3: Моделирование реализации случайного процесса

Промоделируем одну реализацию случайного процесса, управляемого оптимальной стратегией. Это позволит наглядно продемонстрировать поведение системы: динамику энергии, переключения канала и накопление доставленных пакетов.

```python
def simulate_mdp(policy, T_val, e_init=10, c_init=None, seed=42):
    """
    Симуляция одной реализации MDP с оптимальной стратегией.
    
    Параметры:
        policy  - оптимальная стратегия policy[t, e, c]
        T_val   - горизонт
        e_init  - начальная энергия
        c_init  - начальное состояние канала (None = случайное из стац. распр.)
        seed    - зерно ГСЧ
    
    Возвращает словарь с траекториями.
    """
    rng = np.random.default_rng(seed)
    
    if c_init is None:
        c_init = rng.choice([0, 1], p=pi_channel)
    
    # Массивы для хранения траектории
    energy_traj = np.zeros(T_val + 1, dtype=int)
    channel_traj = np.zeros(T_val + 1, dtype=int)
    action_traj = np.zeros(T_val, dtype=int)
    delivered_traj = np.zeros(T_val, dtype=int)
    
    energy_traj[0] = e_init
    channel_traj[0] = c_init
    
    for t in range(T_val):
        e = energy_traj[t]
        c = channel_traj[t]
        
        # Оптимальное управление
        u = policy[t, e, c]
        action_traj[t] = u
        
        # Результат передачи
        if u == 1:
            p_deliver = p_good if c == 1 else p_bad
            delivered_traj[t] = int(rng.random() < p_deliver)
        
        # Переход энергии
        e_base = e - u
        xi = int(rng.random() < p_solar)
        e_new = min(e_base + xi, E_max)
        energy_traj[t + 1] = e_new
        
        # Переход канала
        c_new = rng.choice([0, 1], p=P_ch[c])
        channel_traj[t + 1] = c_new
    
    return {
        'energy': energy_traj,
        'channel': channel_traj,
        'action': action_traj,
        'delivered': delivered_traj,
        'cumulative_delivered': np.cumsum(delivered_traj),
        'T': T_val,
        'e_init': e_init,
        'c_init': c_init
    }
```

```python
# Моделирование для T = 5000
sim = simulate_mdp(policy_5000, T_main, e_init=10, seed=42)

print(f"Начальные условия: e={sim['e_init']}, c={sim['c_init']} "
      f"({'благоприятный' if sim['c_init']==1 else 'неблагоприятный'})")
print(f"Всего попыток передачи: {sim['action'].sum()}")
print(f"Доставлено пакетов: {sim['delivered'].sum()}")
print(f"Доля успешных передач: {sim['delivered'].sum() / max(sim['action'].sum(), 1):.3f}")
print(f"Теоретическое ожидание W₀(e=10, c={sim['c_init']}): "
      f"{W0_5000[10, sim['c_init']]:.2f}")
```

```python
# Визуализация траектории
fig, axes = plt.subplots(4, 1, figsize=(14, 12), sharex=True)

# Показываем первые 500 шагов для наглядности
n_show = 500
t_range = np.arange(n_show)

# 1. Уровень энергии
ax = axes[0]
ax.plot(t_range, sim['energy'][:n_show], color='#20808D', linewidth=0.8)
ax.fill_between(t_range, 0, sim['energy'][:n_show], alpha=0.15, color='#20808D')
ax.set_ylabel('Энергия e')
ax.set_title('Уровень заряда батареи')
ax.set_ylim(0, E_max + 1)
ax.axhline(y=E_max, color='gray', linestyle='--', linewidth=0.5, label=f'$E_{{max}}$={E_max}')
ax.legend(loc='upper right')

# 2. Состояние канала
ax = axes[1]
ax.fill_between(t_range, 0, sim['channel'][:n_show],
                step='post', alpha=0.3, color='#A84B2F')
ax.step(t_range, sim['channel'][:n_show], where='post', color='#A84B2F', linewidth=0.8)
ax.set_ylabel('Канал c')
ax.set_title('Состояние канала связи (0 = неблагоприятный, 1 = благоприятный)')
ax.set_yticks([0, 1])
ax.set_yticklabels(['Неблаг.', 'Благ.'])

# 3. Действия и доставки
ax = axes[2]
transmit_t = t_range[sim['action'][:n_show] == 1]
delivered_t = t_range[sim['delivered'][:n_show] == 1]
failed_t = np.setdiff1d(transmit_t, delivered_t)
ax.bar(delivered_t, 1, width=1, color='#20808D', alpha=0.7, label='Доставлено')
ax.bar(failed_t, 1, width=1, color='#A84B2F', alpha=0.7, label='Не доставлено')
ax.set_ylabel('Передача')
ax.set_title('Попытки передачи и результат')
ax.legend(loc='upper right')
ax.set_yticks([0, 1])

# 4. Кумулятивные доставленные пакеты
ax = axes[3]
ax.plot(np.arange(T_main), sim['cumulative_delivered'], color='#1B474D', linewidth=1)
ax.set_ylabel('Доставлено (всего)')
ax.set_xlabel('Временной слот t')
ax.set_title(f'Накопленное число доставленных пакетов (итого: {sim["delivered"].sum()})')

plt.tight_layout()
plt.savefig('simulation_trajectory.png', dpi=150, bbox_inches='tight')
plt.show()
```

### Множественные симуляции для верификации

Проведём серию независимых симуляций и сравним среднее фактическое число доставленных пакетов с теоретическим значением из функции Беллмана.

```python
# Множественные симуляции для верификации
n_simulations = 200
delivered_counts = []

for i in range(n_simulations):
    sim_i = simulate_mdp(policy_5000, T_main, e_init=10, seed=1000 + i)
    delivered_counts.append(sim_i['delivered'].sum())

delivered_counts = np.array(delivered_counts)

print(f"Результаты {n_simulations} симуляций (T={T_main}, e_init=10):")
print(f"  Среднее число доставленных пакетов: {delivered_counts.mean():.1f}")
print(f"  Стандартное отклонение: {delivered_counts.std():.1f}")
print(f"  95%-доверительный интервал: "
      f"[{delivered_counts.mean() - 1.96*delivered_counts.std()/np.sqrt(n_simulations):.1f}, "
      f"{delivered_counts.mean() + 1.96*delivered_counts.std()/np.sqrt(n_simulations):.1f}]")
print(f"  Мин / Макс: {delivered_counts.min()} / {delivered_counts.max()}")
print(f"  Теоретическое ожидание W₀(10, стац. распр.): "
      f"{pi_channel[0]*W0_5000[10,0] + pi_channel[1]*W0_5000[10,1]:.1f}")

# Гистограмма
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(delivered_counts, bins=25, color='#20808D', edgecolor='white', alpha=0.8)
theoretical = pi_channel[0]*W0_5000[10,0] + pi_channel[1]*W0_5000[10,1]
ax.axvline(theoretical, color='#A84B2F', linewidth=2, linestyle='--',
           label=f'Теория: {theoretical:.1f}')
ax.axvline(delivered_counts.mean(), color='#1B474D', linewidth=2, linestyle='-',
           label=f'Среднее симуляций: {delivered_counts.mean():.1f}')
ax.set_xlabel('Число доставленных пакетов')
ax.set_ylabel('Частота')
ax.set_title(f'Распределение числа доставленных пакетов ({n_simulations} симуляций, T={T_main})')
ax.legend()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('simulation_histogram.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Задание 4: Влияние вероятности подзарядки $p_{\text{solar}}$

Исследуем, как вероятность получения энергии от солнечной батареи влияет на ожидаемое число доставленных пакетов. Варьируем $p_{\text{solar}}$ от 0.05 до 0.5 и для каждого значения решаем задачу ДП.

```python
%%time
# Исследование влияния p_solar
p_solar_values = np.arange(0.05, 0.55, 0.05)
T_analysis = 5000
e_init_analysis = 10

# Результаты: для каждого p_solar храним ожидаемое число пакетов
expected_packets_c0 = []  # начальное состояние канала = 0
expected_packets_c1 = []  # начальное состояние канала = 1
expected_packets_avg = []  # среднее по стационарному распределению

for ps in p_solar_values:
    W0 = solve_bellman(T_analysis, p_solar_val=ps)
    expected_packets_c0.append(W0[e_init_analysis, 0])
    expected_packets_c1.append(W0[e_init_analysis, 1])
    avg = pi_channel[0] * W0[e_init_analysis, 0] + pi_channel[1] * W0[e_init_analysis, 1]
    expected_packets_avg.append(avg)
    print(f"  p_solar = {ps:.2f}: E[доставленных пакетов] = {avg:.1f}")

expected_packets_c0 = np.array(expected_packets_c0)
expected_packets_c1 = np.array(expected_packets_c1)
expected_packets_avg = np.array(expected_packets_avg)
```

```python
# Визуализация
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Левый график: абсолютные значения
ax = axes[0]
ax.plot(p_solar_values, expected_packets_avg, 'o-', color='#20808D',
        linewidth=2, markersize=6, label='Среднее (стац. распр.)')
ax.plot(p_solar_values, expected_packets_c1, 's--', color='#1B474D',
        linewidth=1.5, markersize=5, label='Благоприятный канал (c=1)', alpha=0.7)
ax.plot(p_solar_values, expected_packets_c0, 'D--', color='#A84B2F',
        linewidth=1.5, markersize=5, label='Неблагоприятный канал (c=0)', alpha=0.7)
ax.set_xlabel('Вероятность подзарядки $p_{solar}$')
ax.set_ylabel('Ожидаемое число пакетов')
ax.set_title(f'Ожидаемое число доставленных пакетов (T={T_analysis})')
ax.legend()
ax.grid(True, alpha=0.3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Правый график: средняя скорость доставки (пакетов/слот)
ax = axes[1]
rate = expected_packets_avg / T_analysis
ax.plot(p_solar_values, rate, 'o-', color='#20808D', linewidth=2, markersize=6,
        label='Оптимальная стратегия')

# Теоретические пределы
theoretical_upper = p_solar_values * p_good
theoretical_naive = p_solar_values * (pi_0 * p_bad + pi_1 * p_good)
ax.plot(p_solar_values, theoretical_upper, '--', color='gray', linewidth=1,
        label=f'Предел: $p_{{solar}} \\times p_{{good}}$')
ax.plot(p_solar_values, theoretical_naive, ':', color='#A84B2F', linewidth=1,
        label='Наивная стратегия (без выбора канала)')

ax.set_xlabel('Вероятность подзарядки $p_{solar}$')
ax.set_ylabel('Средняя скорость доставки (пакетов/слот)')
ax.set_title('Эффективность оптимальной стратегии')
ax.legend()
ax.grid(True, alpha=0.3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('p_solar_analysis.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# Количественный анализ
print("Подробный анализ:")
print(f"{'p_solar':>8s}  {'E[пакеты]':>10s}  {'пакет/слот':>11s}  "
      f"{'пакет/энергию':>14s}  {'% от предела':>12s}")
print("-" * 60)
for i, ps in enumerate(p_solar_values):
    rate_i = expected_packets_avg[i] / T_analysis
    efficiency = rate_i / ps if ps > 0 else 0  # доставленных на единицу энергии
    pct_limit = rate_i / (ps * p_good) * 100  # % от теоретического предела
    print(f"{ps:8.2f}  {expected_packets_avg[i]:10.1f}  {rate_i:11.4f}  "
          f"{efficiency:14.4f}  {pct_limit:11.1f}%")
```

### Выводы по влиянию $p_{\text{solar}}$

1. **При малых $p_{\text{solar}}$** (0.05–0.20) зависимость **практически линейная**: удвоение вероятности подзарядки почти удваивает число доставленных пакетов. Это объясняется тем, что энергия — основное ограничение, и её увеличение линейно масштабирует число возможных передач.

2. **При больших $p_{\text{solar}}$** (>0.35) рост **замедляется** — кривая становится субпроизводственной. Причины:
   - Ёмкость батареи $E_{\max} = 20$ ограничивает накопление: лишняя энергия при полной батарее теряется.
   - Канал находится в благоприятном состоянии лишь ~40% времени, и при избытке энергии сенсор вынужден передавать и в неблагоприятном канале (с $p_{\text{bad}} = 0.45$).

3. **Эффективность использования энергии** (пакетов на единицу потреблённой энергии) **снижается** при увеличении $p_{\text{solar}}$, т.к. всё большая доля передач приходится на неблагоприятный канал.

4. Оптимальная стратегия **существенно лучше наивной** (передача в любом канале): при $p_{\text{solar}} = 0.2$ оптимальная стратегия доставляет на ~50% больше пакетов за счёт «ожидания» благоприятного канала.

## Итоги

В данной работе решена задача **стохастического динамического программирования** для оптимального управления передачей данных автономного беспроводного сенсора.

### Основные результаты:

1. **Оптимальная стратегия** найдена методом обратной прогонки уравнения Беллмана. Она имеет **пороговый характер**: для каждого состояния канала существует минимальный уровень энергии, начиная с которого оптимально передавать. В стационарном режиме сенсор передаёт только в благоприятном канале (за исключением полной батареи).

2. **Стратегии для различных горизонтов** ($T=1000$, $T=2000$, $T=5000$) совпадают в стационарной части и различаются лишь вблизи конца горизонта (последние ~30 шагов), где стратегия становится более «агрессивной».

3. **Симуляция подтверждает** теоретические результаты: среднее число доставленных пакетов по множеству симуляций хорошо согласуется с ожиданием $W_0$.

4. **Зависимость от $p_{\text{solar}}$** близка к линейной при малых значениях и замедляется при больших (из-за ограниченной ёмкости батареи и необходимости передавать в неблагоприятном канале).

### Методологические замечания:

- Задача является классическим **марковским процессом принятия решений** (MDP) с конечным горизонтом.
- Уравнение Беллмана для стохастического случая содержит **математическое ожидание** по случайным переходам (в отличие от детерминированного ДП).
- Метод обратной прогонки гарантирует нахождение **глобально оптимальной** стратегии благодаря принципу оптимальности Беллмана.
- Аналогия с задачей о садовнике: там управление — внесение удобрений, здесь — передача данных; там состояние — качество почвы, здесь — энергия + состояние канала.
